```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
class DataRead1(object):
    def __init__(self, file_path):
        self.file_path = pd.read_csv(file_path)

    def file_read(self):
        return self.file_path

    def RowSelect(self):
        row_1 = int(input('Enter index of the row 1: '))
        row_2 = int(input('Enter index of row 2: '))
        row_12 = self.file_path.iloc[[row_1, row_2], :]
        return row_12

class DataRead2(DataRead1):
    def correlation(self):
        corr = self.file_path.corr()
        sns.heatmap(corr)

    def RowSelect(self):
        row_1 = int(input('Enter index of the row 1: '))
        row_2 = int(input('Enter index of row 2: '))
        row_3 = int(input('Enter index of row 3: '))
        row_123 = self.file_path.iloc[[row_1, row_2, row_3], :]
        return row_123
```
